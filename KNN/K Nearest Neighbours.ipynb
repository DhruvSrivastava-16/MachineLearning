```python
def eucladian_dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))
```

```python
def KNN(X,Y,querypt,K=5):
    V = []
    m = X.shape[0]
    for i in range(m):
        d = eucladian_dist(X[i],querypt)
        V.append((d,Y[i]))
    
    V = sorted(V)
    V = V[:K]
    V = np.array(V)
    V_freq = np.unique(V[:,1],return_counts=True)
    #print(V_freq)
    index = V_freq[1].argmax()
    pred = V_freq[0][index]
    #print(V[0][1])
    return pred
```

```python
pred = KNN(x,y,querypt)
print('it mostly is of the class: ',pred)
```
